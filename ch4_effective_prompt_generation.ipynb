```python
from google.colab import userdata
openai_api_key = userdata.get('openai_api_key')
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.schema import HumanMessage, SystemMessage, AIMessage

chat = ChatOpenAI(
    model_name = 'gpt-3.5-turbo',
    openai_api_key = openai_api_key,
    temperature = 0.5
)
```

```python
## Simple Chatbot
## To run code --> first get open api_key from openai platform

base_conversation = [
    SystemMessage(content="You are a chef, you want to ask me some things to find out my favorite Asian dish."),
]

while True:
    ai_response = chat(base_conversation)
    print("AI: ", ai_response.content)
    base_conversation.append(AIMessage(content = ai_response.content))
    _content = input("Your Answer: ")
    if _content == "exit":
        break

    user_message = HumanMessage(content = _content)
    base_conversation.append(user_message)

human_final_messang = HumanMessage(content="And now. Which plate do you think I like the most?")
base_conversation.append(human_final_messang)
final_response = chat(base_conversation)
print("AI: ", final_response.content)
```

## Chatting
* System: Context or Background information

* Human: Message from the user

* AI: Message from the LLM

```python
from langchain.llms import OpenAI
from langchain.chat_models import ChatOpenAI
from langchain.schema import HumanMessage, SystemMessage, AIMessage


chat = ChatOpenAI(
    model_name = 'text-davinci-003',
    openai_api_key = openai_api_key,
    temperature = 0.5
)

```

```python
system_message = SystemMessage(
    content = "You are a helpful assistant and answer concisely."
)

human_message = HumanMessage(
    content = "Explain how network firewalls work. "
)

response = chat(
    [
        system_message,
        human_message
    ]
)

print(response.content)  # you must have openai api_key to get response
```
