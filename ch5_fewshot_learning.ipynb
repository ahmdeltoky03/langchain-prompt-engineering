## Basic FewShots

```python
from langchain.prompts.prompt import PromptTemplate
from langchain.prompts.few_shot import FewShotPromptTemplate
```

```python
demo_message = """
Country Name: Egypt | Capital Name: Cairo
Country Name: Saudi Arabia | Capital Name: Riyadh
Country Name: Singapore | Capital Name: Singapore
Country Name: Armenia |
""".strip()

print(llm(demo_message))
```

```python
demo_template = PromptTemplate (
    template = "country name: {country} | capital name: {capital}",
    input_variables = ["country", "capital"]
)

examples = [
    {"country": "Egypt", "capital": "Cairo"},
    {"country": "Saudi Arabia", "capital": "Riyadh"},
    {"country": "Singapore", "capital": "Singapore"},
]

prompt = FewShotPromptTemplate(
    example_prompt = demo_template,
    examples = examples,
    suffix = "country name: {country} |",
    input_variables = ["country"]
)
```

```python
country = "Italy"
message = prompt.format(country=country)
## few shot message
# print(message)
```

```python
print(llm(message))
```

## FewShot Learning

```python
problem_template = "when I was five years old my sister was twice than we. I am sixty years old now, what age of my sister ? "

print( llm ( problem_template ) )
```

```python
problem_template = "I have three pen and my sister have three times than me plus one. How many pen does It have ? "

print( llm ( problem_template ) )
```

```python
examples = [
    {
        "question": "When I was seven, my sister was twice my age. Now I am seventy years old, what age of my sister ?",
        "answer": "\n".join([
            "We will followup some questions to get the answer.",
            "Follow up: How old was your sister when you were seven?",
            "Intermediate answer: Twice, which mean 14 years.",
            "Follow up: What is the difference between your age and your sister's age?",
            "Intermediate answer: 14 years - 7 years = 7 years.",
            "Follow up: When you were seventy years old, how old would your sister be?",
            "Intermediate answer: my age (70) + The difference between me and my sister's age (7 years)",
            "Final Answer: 77 years."
        ])
    }
]

template = PromptTemplate(
    template = "Question: {question}\nAnswer: {answer}",
    innput_variables = ["question", "answer"]
)


prompt = FewShotPromptTemplate(
    example_prompt = template,
    examples = examples,
    suffix = "Question: {question}'\nAnswer",
    input_variables=['question']
)
```

```python
new_problem = "I have three pen and my sister have three times than me plus one. How many pen does It have ? "
message = prompt.format(question=new_problem)

print(message)
```

```python
print(llm(message))
```
